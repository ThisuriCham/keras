```
eng_words = []
sia_words = []
eng_chars = set()
sia_chars = set()
nb_samples = 1000

# Process english and sinhala words
for word in range(nb_samples):
    
    eng_word = str(words[word]).split('\t')[0]
    
    # Append '\t' for start of the word and '\n' to signify end of the word
    sia_word = '\t' + str(words[word]).split('\t')[1] + '\n'
    eng_words.append(eng_word)
    sia_words.append(sia_word)
    
    for ch in eng_word:
        if (ch not in eng_chars):
            eng_chars.add(ch)
            
    for ch in sia_word:
        if (ch not in sia_chars):
            sia_chars.add(ch)
```

```
sia_chars = sorted(list(sia_chars))
eng_chars = sorted(list(eng_chars))
```

```
# dictionary to index each english character - key is index and value is english character
eng_index_to_char_dict = {}

# dictionary to get english character given its index - key is english character and value is index
eng_char_to_index_dict = {}

for k, v in enumerate(eng_chars):
    eng_index_to_char_dict[k] = v
    eng_char_to_index_dict[v] = k
    
echar_index = dict((c, i) for i, c in enumerate(eng_chars))
print(echar_index)
```

```
# dictionary to index each sinhala character - key is index and value is sinhala character
sia_index_to_char_dict = {}

# dictionary to get sinhala character given its index - key is sinhala character and value is index
sia_char_to_index_dict = {}

for k, v in enumerate(sia_chars):
   sia_index_to_char_dict[k] = v
   sia_char_to_index_dict[v] = k
    
schar_index = dict((c, i) for i, c in enumerate(sia_chars))
print(schar_index)
```

```
max_len_eng_word = max([len(word) for word in eng_words])
max_len_sia_word = max([len(word) for word in sia_words])
```

```
print(max_len_eng_word)
print(max_len_sia_word)
```

```
tokenized_eng_words = np.zeros(shape = (nb_samples,max_len_eng_word,len(eng_chars)), dtype='float32')
tokenized_sia_words = np.zeros(shape = (nb_samples,max_len_sia_word,len(sia_chars)), dtype='float32')
target_data = np.zeros((nb_samples, max_len_sia_word, len(sia_chars)),dtype='float32')
```

```
# Vectorize the english and sinhala sentences

for i in range(nb_samples):
    for k,ch in enumerate(eng_words[i]):
        tokenized_eng_words[i,k,eng_char_to_index_dict[ch]] = 1
        
    for k,ch in enumerate(sia_words[i]):
        tokenized_sia_words[i,k,sia_char_to_index_dict[ch]] = 1

        # decoder_target_data will be ahead by one timestep and will not include the start character.
        if k > 0:
            target_data[i,k-1,sia_char_to_index_dict[ch]] = 1
```

```
# Encoder model

encoder_input = Input(shape=(None,len(eng_chars)))
encoder_LSTM = LSTM(256,return_state = True)
encoder_outputs, encoder_h, encoder_c = encoder_LSTM (encoder_input)
encoder_states = [encoder_h, encoder_c]
```

```
# Decoder model

decoder_input = Input(shape=(None,len(sia_chars)))
decoder_LSTM = LSTM(256,return_sequences=True, return_state = True)
decoder_out, _ , _ = decoder_LSTM(decoder_input, initial_state=encoder_states)
decoder_dense = Dense(len(sia_chars),activation='softmax')
decoder_out = decoder_dense (decoder_out)
```

```
model = Model(inputs=[encoder_input, decoder_input],outputs=[decoder_out])

# Run training
model.compile(optimizer='rmsprop', loss='categorical_crossentropy')
model.fit(x=[tokenized_eng_words,tokenized_sia_words], 
          y=target_data,
          batch_size=64,
          epochs=30,
          validation_split=0.2)
```

```
# Inference models for testing

# Encoder inference model
encoder_model_inf = Model(encoder_input, encoder_states)

# Decoder inference model
decoder_state_input_h = Input(shape=(256,))
decoder_state_input_c = Input(shape=(256,))
decoder_input_states = [decoder_state_input_h, decoder_state_input_c]

decoder_out, decoder_h, decoder_c = decoder_LSTM(decoder_input, 
                                                 initial_state=decoder_input_states)

decoder_states = [decoder_h , decoder_c]

decoder_out = decoder_dense(decoder_out)

decoder_model_inf = Model(inputs=[decoder_input] + decoder_input_states,
                          outputs=[decoder_out] + decoder_states )
```

```
def decode_seq(inp_seq):
    
    # Initial states value is coming from the encoder 
    states_val = encoder_model_inf.predict(inp_seq)
    
    target_seq = np.zeros((1, 1, len(sia_chars)))
    target_seq[0, 0, sia_char_to_index_dict['\t']] = 1
    
    translated_sent = ''
    stop_condition = False
    
    while not stop_condition:
        
        decoder_out, decoder_h, decoder_c = decoder_model_inf.predict(x=[target_seq] + states_val)
        
        max_val_index = np.argmax(decoder_out[0,-1,:])
        sampled_sia_char = sia_index_to_char_dict[max_val_index]
        translated_sent += sampled_sia_char
        
        if ( (sampled_sia_char == '\n') or (len(translated_sent) > max_len_sia_sent)) :
            stop_condition = True
        
        target_seq = np.zeros((1, 1, len(sia_chars)))
        target_seq[0, 0, max_val_index] = 1
        
        states_val = [decoder_h, decoder_c]
        
    return translated_sent
```
